```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
def losses(path):
    try:
        with open(path + "/loss_curve", "rb") as f:
            l = np.array(pickle.load(f ))
        return l[:]
    except:
        with open(path + "/loss.pickle", "rb") as f:
            l = np.array(pickle.load(f ))
        return l[:]
```

```python

for c, name in zip(curves, fnames):
    curve = c[:, 2]
    curve = curve.reshape(-1, 50)
    curve = np.sum(curve, axis=1) / 50
    plt.plot(curve, label=name[15:], alpha=.9)
plt.legend()
```
